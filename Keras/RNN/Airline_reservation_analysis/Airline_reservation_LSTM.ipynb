# Part 1: LSTM for Regression # 

We will be using the same data loading techniques from the MLP notebook.

```python
'''
Same loading techniques as we used for the Airlines MLP notebook
'''
# LSTM for international airline passengers problem with regression framing
import numpy
import matplotlib.pyplot as plt
import pandas
import math
from keras.models import Sequential
from keras.layers import Dense
# LSTM import
from keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error

# Convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return numpy.array(dataX), numpy.array(dataY)

# Fix random seed for reproducibility
numpy.random.seed(7)

# Load the dataset
dataframe = pandas.read_csv('international-airline-passengers.csv', usecols=[1], engine='python', skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')
```

```python
'''
LSTMs are sensitive to the scale of the input data, specifically when the sigmoid or tanh activation functions are
used. We will rescale the data to 0-1, normalize the data.
'''
# normalize the dataset
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)
```

```python
'''
Split the data into training(first 67%) and testing sets(last 33%). Since the values are time dependant, we cannot
shuffle the data.
'''
# split into train and test sets
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size

train, test = dataset[0:train_size,:], \
                dataset[train_size:len(dataset),:]
```

```python
'''
Using the helper function, reshape the data.
'''
# reshape into X=t and Y=t+1
look_back = 1
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)
```

```python
'''
Check structure after using helper function
'''
print trainX[:5]
```

```python
'''
LSTMs expect the input data to be of a specific format: [samples, time_steps, features]. Our data is of the form
[samples, features]
'''
# Reshape input to be [samples, time steps, features]

# Training data
trainX = numpy.reshape(trainX, (trainX.shape[0], 
                                1, 
                                trainX.shape[1]))

# Testing data
testX = numpy.reshape(testX, (testX.shape[0], 
                              1, 
                              testX.shape[1]))
```

```python
'''
Sanity check the training and testing sets
'''
print trainX[:5]
print testX[:5]
```

```python
'''
Create LSTM model

- 1 visible layer with 1 input
- 4 LSTM blocks or neurons
- 1 output layer that makes a single value prediction

Default sigmoid activation function for the LSTM memory blocks, 50 epochs and a batch size of 1.
'''
# Create and fit the LSTM network
model = Sequential()
model.add(LSTM(4, 
               input_dim = look_back))
model.add(Dense(1))
model.compile(loss='mean_squared_error', 
              optimizer='adam')
model.fit(trainX, 
          trainY, 
          nb_epoch=50, 
          batch_size=1, 
          verbose=2)
```

```python
'''
Make predictions
'''
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)
```

```python
'''
Re format the predictions to compare against original values for performance evaluation
'''
# Invert predictions on training set
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])

# Invert predictions on test set
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])
```

```python
'''
Performance evaluation
'''
# Calculate root mean squared error on training set
trainScore = math.sqrt(mean_squared_error(trainY[0], 
                                          trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))


# Calculate root mean squared error on testing set
testScore = math.sqrt(mean_squared_error(testY[0], 
                                         testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))
```

```python
'''
Shift the values and plot the data
'''
# Shift train predictions for plotting
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

# Shift test predictions for plotting
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict

# Plot baseline and predictions
%pylab inline
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

# Part 2: LSTM for regression using the window method #

* We will use multiple recent time steps to make the prediction for the next time step. This is called window and the size of the window can be tuned based on the problem.

```python
'''
Window regression framing
'''
# Load the dataset
dataframe = pandas.read_csv('international-airline-passengers.csv', 
                            usecols=[1], 
                            engine='python', 
                            skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')

'''
Normalize the dataset
'''
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

'''
Split into train and test sets
'''
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

'''
Reshape into X = t and Y = t+1

Use previous 3 time stamps to make prediction for next time stamp
'''
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

'''
Reshape input to be [samples, time steps, features]

samples = number of rows
time steps = 1
features = number of columns
'''
trainX = numpy.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
testX = numpy.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

'''
Create and fit the LSTM network
'''
model = Sequential()
model.add(LSTM(4, 
               input_dim = look_back))
# 1 hidden layer
model.add(Dense(1))
model.compile(loss='mean_squared_error', 
              optimizer='adam')
model.fit(trainX, 
          trainY, 
          nb_epoch=50, 
          batch_size=1, 
          verbose=2)
'''
Make predictions
'''
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

'''
Invert predictions
'''
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

'''
Calculate root mean squared error
'''
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))

'''
Shift train predictions for plotting
'''
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

'''
Shift test predictions for plotting
'''
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict

'''
Plot baseline and predictions
'''
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

# Part 3: LSTM for regression with time steps #

* Some sequence problems have a varied number of time steps per sample.
* For eg: measurements of a machine leading up to a failure/surge. Each incident would be a **sample**. the observations that lead up to the event would be the **time steps**, and variables would be the ** features **.
* Time steps can be a way to phrase the problem.
* Instead of using the past time steps as separate input features, we can use them as time steps for the one input feature. 
* The steps are the same as in part 2, ** except when we reshape the data we set the columns to be the time steps dimension and change the feature dimension to 1. **.

```python
'''
Window regression framing
'''
# Load the dataset
dataframe = pandas.read_csv('international-airline-passengers.csv', 
                            usecols=[1], 
                            engine='python', 
                            skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')

'''
Normalize the dataset
'''
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

'''
Split into train and test sets
'''
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

'''
Reshape into X = t and Y = t+1

Use previous 3 time stamps to make prediction for next time stamp
'''
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

'''
*************************************** CHANGE FROM PART 1 IN THIS STEP ***************************************
Reshape input to be [samples, time steps, features]

samples = number of rows
time steps = number of columns
features = 1
'''
# trainX = numpy.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
# testX = numpy.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

trainX = numpy.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = numpy.reshape(testX, (testX.shape[0], testX.shape[1], 1))

'''
*************************************** CHANGE FROM PART 1 IN THIS STEP ***************************************
Create and fit the LSTM network
'''
model = Sequential()
model.add(LSTM(4, 
               input_dim = 1)) # in part input_dim = look_back
# 1 hidden layer
model.add(Dense(1))
model.compile(loss='mean_squared_error', 
              optimizer='adam')
model.fit(trainX, 
          trainY, 
          nb_epoch=50, 
          batch_size=1, 
          verbose=2)
'''
Make predictions
'''
trainPredict = model.predict(trainX)
testPredict = model.predict(testX)

'''
Invert predictions
'''
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

'''
Calculate root mean squared error
'''
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))

'''
Shift train predictions for plotting
'''
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

'''
Shift test predictions for plotting
'''
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict

'''
Plot baseline and predictions
'''
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

# Part 3: LSTM with memory between batches #

* LSTM networks has memory which is capable of remembering across long sequences.
* Normally the state within the network is reset after each training batch when fitting the model as well as each call to model.predict() or model.evaluate()
* When the internal state of the LSTM network is cleared in Keras by making the LSTM layer stateful. This means that ** it can build over the entire training sequence and even maintain that state if needed to make predictions. **

It requires the following:

* Training data not be shuffled when fitting the network.
* Explicit resetting of the network state after each exposure to the training data(after each epoch) by calls to model.reset_states()
* We must create our own outer loop of epochs and within each opoch call model.fit() and model.reset_states()

When the LSTM layer is constructed, the ** statefull ** parameter must be set to True and instead of specifying input dimensions we must hard code the **number of samples in a batch, number of time steps in a sample, number of features in a time step** by using the batch_input_shape parameter.

```python
'''
LSTM with memory

Identical to Part 2 unless specified
'''
# Load the dataset
dataframe = pandas.read_csv('international-airline-passengers.csv', 
                            usecols=[1], 
                            engine='python', 
                            skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')

'''
Normalize the dataset
'''
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

'''
Split into train and test sets
'''
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

'''
Reshape into X = t and Y = t+1

Use previous 3 time stamps to make prediction for next time stamp
'''
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

'''
Reshape input to be [samples, time steps, features]

samples = number of rows
time steps = number of columns
features = 1
'''
# trainX = numpy.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
# testX = numpy.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

trainX = numpy.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = numpy.reshape(testX, (testX.shape[0], testX.shape[1], 1))

'''
*************************************** CHANGE FROM PART 2 IN THIS STEP ***************************************
Create and fit the LSTM network
'''
batch_size = 1
model = Sequential()
model.add(LSTM(4, 
               batch_input_shape = (batch_size, look_back, 1), 
               stateful=True))
model.add(Dense(1))
model.compile(loss='mean_squared_error', 
              optimizer='adam')
for i in range(100):
    model.fit(trainX, 
              trainY, 
              nb_epoch=1, 
              batch_size=batch_size, 
              verbose=2, 
              shuffle=False)
    model.reset_states()
'''
*************************************** CHANGE FROM PART 2 IN THIS STEP ***************************************
Make predictions
'''
trainPredict = model.predict(trainX, 
                             batch_size = batch_size)
testPredict = model.predict(testX, 
                            batch_size = batch_size)

'''
Invert predictions
'''
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

'''
Calculate root mean squared error
'''
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))

'''
Shift train predictions for plotting
'''
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

'''
Shift test predictions for plotting
'''
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict

'''
Plot baseline and predictions
'''
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```

# Part 4: Stacked LSTMs with memory between batches #

* An LSTM layer prior to each subsequent LSTM layer must return the sequence, this is done using ** return_sequences **

```python
'''
LSTM with memory

Identical to Part 3 unless specified
'''
# Load the dataset
dataframe = pandas.read_csv('international-airline-passengers.csv', 
                            usecols=[1], 
                            engine='python', 
                            skipfooter=3)
dataset = dataframe.values
dataset = dataset.astype('float32')

'''
Normalize the dataset
'''
scaler = MinMaxScaler(feature_range=(0, 1))
dataset = scaler.fit_transform(dataset)

'''
Split into train and test sets
'''
train_size = int(len(dataset) * 0.67)
test_size = len(dataset) - train_size
train, test = dataset[0:train_size,:], dataset[train_size:len(dataset),:]

'''
Reshape into X = t and Y = t+1

Use previous 3 time stamps to make prediction for next time stamp
'''
look_back = 3
trainX, trainY = create_dataset(train, look_back)
testX, testY = create_dataset(test, look_back)

'''
Reshape input to be [samples, time steps, features]

samples = number of rows
time steps = number of columns
features = 1
'''
# trainX = numpy.reshape(trainX, (trainX.shape[0], 1, trainX.shape[1]))
# testX = numpy.reshape(testX, (testX.shape[0], 1, testX.shape[1]))

trainX = numpy.reshape(trainX, (trainX.shape[0], trainX.shape[1], 1))
testX = numpy.reshape(testX, (testX.shape[0], testX.shape[1], 1))

'''
*************************************** CHANGE FROM PART 3 IN THIS STEP ***************************************
Create and fit the LSTM network
'''
batch_size = 1
model = Sequential()
model.add(LSTM(4, 
               batch_input_shape=(batch_size, look_back, 1), 
               stateful=True, 
               return_sequences=True))
model.add(LSTM(4, 
               batch_input_shape=(batch_size, look_back, 1), 
               stateful=True))
model.add(Dense(1))
model.compile(loss='mean_squared_error', 
              optimizer='adam')
for i in range(100):
    model.fit(trainX, 
              trainY, 
              nb_epoch=1, 
              batch_size=batch_size, 
              verbose=2, 
              shuffle=False)
    model.reset_states()
'''
Make predictions
'''
trainPredict = model.predict(trainX, 
                             batch_size = batch_size)
testPredict = model.predict(testX, 
                            batch_size = batch_size)

'''
Invert predictions
'''
trainPredict = scaler.inverse_transform(trainPredict)
trainY = scaler.inverse_transform([trainY])
testPredict = scaler.inverse_transform(testPredict)
testY = scaler.inverse_transform([testY])

'''
Calculate root mean squared error
'''
trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
print('Train Score: %.2f RMSE' % (trainScore))
testScore = math.sqrt(mean_squared_error(testY[0], testPredict[:,0]))
print('Test Score: %.2f RMSE' % (testScore))

'''
Shift train predictions for plotting
'''
trainPredictPlot = numpy.empty_like(dataset)
trainPredictPlot[:, :] = numpy.nan
trainPredictPlot[look_back:len(trainPredict)+look_back, :] = trainPredict

'''
Shift test predictions for plotting
'''
testPredictPlot = numpy.empty_like(dataset)
testPredictPlot[:, :] = numpy.nan
testPredictPlot[len(trainPredict)+(look_back*2)+1:len(dataset)-1, :] = testPredict

'''
Plot baseline and predictions
'''
plt.plot(scaler.inverse_transform(dataset))
plt.plot(trainPredictPlot)
plt.plot(testPredictPlot)
plt.show()
```
