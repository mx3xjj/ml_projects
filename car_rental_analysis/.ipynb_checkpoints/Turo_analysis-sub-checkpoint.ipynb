# Part 1: Which factors seem to be most important in driving total # of reservations for our vehicles? #

## Step 1: Data preprocessing ##

```python
import pandas as pd
import numpy as np

# Read CSV into a dataframe
reservations_df = pd.read_csv('reservations_(5).csv')
reservations_df.columns = ['vehicle_id', 'reservation_type']
reservations_df = reservations_df.sort_values(by = ['vehicle_id']).reset_index(drop = True)
reservations_df.head(10)
```

```python
# Data preprocessing
reservations_df_gd = pd.get_dummies(reservations_df['reservation_type']).astype(np.int8) \
                   .rename(columns=lambda x: 'reservation_type_' + str(x))

reservations_df = pd.concat([reservations_df, reservations_df_gd], axis=1)
reservations_df = reservations_df.drop('reservation_type', axis = 1)
```

```python
# Number of reservations for each vehicle by reservation type
reservations_df = reservations_df.groupby(["vehicle_id"]).sum().reset_index()

# Create new column with total reservations for each vehicle across all types
reservations_df['total_reservations'] = reservations_df['reservation_type_1'] + reservations_df['reservation_type_2']\
+ reservations_df['reservation_type_3']

# Preview data
reservations_df.head(20)
```

```python
# Now we start with the second CSV file containing information about the vehicle features
vehicles_df = pd.read_csv('vehicles_(6).csv')
vehicles_df.head(10)
```

```python
# Merge the two dataframes on vehicle id
merged_df = pd.merge(vehicles_df, reservations_df, on='vehicle_id')

# Write to csv to save a copy
merged_df.to_csv('merged_df.csv')
```

```python
# Preview final version of dataset we will be doing our analysis with
merged_df.head(10)
```

## Step 2: Setting the feature and target variables ##

```python
# Target column 
target_col = ['total_reservations']

# Feature columns
feature_cols = ['technology', 
                'actual_price', 
                'recommended_price', 
                'num_images', 
                'street_parked', 
                'description']

# Drop the target column 
features = merged_df[feature_cols]
target = merged_df[target_col]
print '**Sample Feature Column values**'
print features.head()
print '**Sample Target Column values**'
print target.head()
```

## Step 3: Outlier detection ##

```python
# For each feature finding the data points with extreme high or low values based on Tukey's outlier detection method
index_to_outliers = {}

# Checking for outliers in the following columns.
outlier_cols = ['actual_price', 
                'recommended_price', 
                'num_images', 
                'description', 
                'reservation_type_1', 
                'reservation_type_2', 
                'reservation_type_3']

for feature in outlier_cols:
    # Calculating Q1 (25th percentile of the data) for the given feature
    Q1 = merged_df[feature].quantile(0.25)
    
    # Calculating Q3 (75th percentile of the data) for the given feature
    Q3 = merged_df[feature].quantile(0.75)
    
    # Use the interquartile range to calculate an outlier step (1.5 times the interquartile range)
    step = step = 1.5 * (Q3 - Q1)
    
    # Display the outliers
    print "Data points considered outliers for the feature '{}':".format(feature)
    outliers = merged_df[~((merged_df[feature] >= Q1 - step) & (merged_df[feature] <= Q3 + step))] 
    print outliers
    
    for i in outliers.index.tolist():
        if i in index_to_outliers:
            index_to_outliers[i] += 1
        else:
            index_to_outliers[i] = 1
    
filtered_dict = {k:v for k,v in index_to_outliers.iteritems() if v > 1}    
```

```python
for key, value in filtered_dict.iteritems():
    print "The datapoint {} has outliers for {} features".format(key, value)
```

```python
# The indices for data points I may remove.
outliers  = [filtered_dict.keys()]
print outliers
```

## Step 4: Splitting and scaling the data ##

### Step 4.1 ###

```python
from sklearn.cross_validation import train_test_split

# X -> features, y -> target
X, y = features.values, target.values

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.3, random_state=0)
```

### Step 4.2 ###

```python
# Standardize the values
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()
X_train_std = stdsc.fit_transform(X_train)
X_test_std = stdsc.transform(X_test)
```


## Step 5: Feature Selection ##

### Step 5.1 ###

```python
# Sequential Backward Selection
from itertools import combinations
from sklearn.base import clone
from sklearn.cross_validation import train_test_split
from sklearn.metrics import r2_score
"""
Sequential Backward Selection:
Sequentially remove features from the full feature set until the new feature set has the desired number 'k' of 
features. At each stage we eliminate the the feature that causes the least performance loss after removal.
"""
class SequentialBackwardSelection():
    # Setting the scoring parameter to mean squared error 
    # k features -> to define the number of top features we want to select
    def __init__(self, 
                 estimator, 
                 k_features, 
                 scoring = r2_score,
                 test_size=0.25, 
                 random_state=1):

        self.scoring = scoring
        self.estimator = clone(estimator)       
        self.k_features = k_features # in our case, k = 3
        self.test_size = test_size
        self.random_state = random_state

    def fit(self, X, y):
        
        X_train, X_test, y_train, y_test = \
            train_test_split(X, y, test_size=self.test_size,
                             random_state=self.random_state)
        # dim is set to the total number of features
        dim = X_train.shape[1]
        self.indices_ = tuple(range(dim))
        self.subsets_ = [self.indices_]
        # r2
        score = self._calc_score(X_train, y_train, 
                                 X_test, y_test, self.indices_)
        self.scores_ = [score]
        
        # continues until the feature set has the desired dimensionality a.k.a k
        while dim > self.k_features:
            scores = []
            subsets = []
            for p in combinations(self.indices_, r = dim - 1):
                score = self._calc_score(X_train, y_train, 
                                         X_test, y_test, p)
                scores.append(score)
                subsets.append(p)
            # best is the index of the highest score
            best = np.argmax(scores)
            self.indices_ = subsets[best]
            self.subsets_.append(self.indices_)
            dim -= 1
            # append the highest score
            self.scores_.append(scores[best])
        # last element in list has the highest score
        self.k_score_ = self.scores_[-1]

        return self

    def transform(self, X):
        return X[:, self.indices_]
    
    
    def _calc_score(self, X_train, y_train, X_test, y_test, indices):
        self.estimator.fit(X_train[:, indices], y_train)
        y_pred = self.estimator.predict(X_test[:, indices])
        score = self.scoring(y_test, y_pred)
        return score
```

### Step 5.2 ###

```python
from sklearn import linear_model
import time
'''
Generally, when you have many small/medium sized effects you should go with ridge. 
If you have only a few variables with a medium/large effect, go with lasso.
'''
lr = linear_model.Ridge(alpha = 1.0)

# selecting features
sbs = SequentialBackwardSelection(lr, k_features = 1)
print 'Fitting model....'
start_time = time.time()
sbs.fit(X_train_std, y_train)
print "\nTime taken to fit the model in seconds: ", time.time() - start_time

# Performance of feature subsets
k_feat = [len(k) for k in sbs.subsets_]

# Getting the top features by sequential removal of 1 feature at a time
feature_subsets = list(sbs.subsets_)
print "\nEliminating the the features that causes the least performance loss after removal at each stage..",\
feature_subsets

print "\nAt each iteration, one feature is removed as follows, until we reach the top feature."
count = 1
for i in xrange(len(feature_subsets)):
    print '\nIteration number: ', count
    count += 1
    for x in feature_subsets[i]:    
        print features.columns[x]
        
```

```python
# Testing model using all features

# Fit the model
lr.fit(X_train_std, y_train)

# Default scoring for Ridge is r^2
training_score = lr.score(X_train_std, y_train) * 100
testing_score = lr.score(X_test_std, y_test) * 100
print 'Training score: {}'.format(training_score)
print 'Test score: {}'.format(testing_score)
```

```python
# Testing model using top 3 features we found Sequential Back Propagation
feature_subsets_3 = list(sbs.subsets_[3])

# Fit the model
lr.fit(X_train_std[:, feature_subsets_3], y_train)

# Default scoring for Ridge is r^2
print 'Training score:', lr.score(X_train_std[:, feature_subsets_3], y_train) * 100
print 'Testing  score:', lr.score(X_test_std[:, feature_subsets_3], y_test) * 100
```

### Step 5.3 ###

```python
# plotting performance of feature subsets
import matplotlib.pyplot as plt
%matplotlib inline  

k_feat = [len(k) for k in sbs.subsets_]
plt.plot(k_feat, 
         sbs.scores_, 
         marker='o')
plt.xlim([0,7])
plt.ylabel('R2 Score')
plt.xlabel('Number of features')
plt.grid()
plt.tight_layout()
plt.show()
```
