# Converting data to time series #

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def convert_data_to_timeseries(input_file, column, verbose = True):
    # Load the input file
    data = np.loadtxt(input_file, 
                      delimiter=',')

    # Extract the start date which is in the first row
    # First row first column + First row second column
    start_date = str(int(data[0,0])) + '-' + str(int(data[0,1]))
    
    # Extract the end date which is the last row
    # Last row first column + Last row second column
    end_date = str(int(data[-1,0] + 1)) + '-' + str(int(data[-1,1] % 12 + 1))

    if verbose:
        print "\nStart date =", start_date
        print "End date =", end_date

    # Create a date sequence with monthly intervals
    dates = pd.date_range(start_date, 
                          end_date, 
                          freq = 'M')

    # Convert the data into time series data
    data_timeseries = pd.Series(data[:,column], 
                                index=dates)

    if verbose:
        print "\nTime series data:\n", data_timeseries[:10]

    return data_timeseries

if __name__=='__main__':
    # Input file containing data
    '''
    Data of the form:
    
    1940,1,98.96,64.81
    1940,2,10.89,71.38
    1940,3,6.22,14.39
    1940,4,67.85,99.24
    1940,5,31.7,5.09
    
    Format
    Year, Month, data, data
    '''
    input_file = 'data_timeseries.txt'

    # Load input data
    column_num = 2
    data_timeseries = convert_data_to_timeseries(input_file, 
                                                 column_num)

    # Plot the time series data
    data_timeseries.plot()
    plt.title('Input data')

    plt.show()
```

# Hidden Markov Model(generative model) #

HMM are generative models that can generate data once they learn the underlying structure. 

They cannot discriminate between classes in their base form. Discriminative models can learn to discriminate between classes but cannot generate data.

**HMMs assume that the current output is statistically independant of the previous outputs. **

```python
# pip install hmmlearn
import datetime
from hmmlearn.hmm import GaussianHMM


# Load data from input file
'''
Data of the form: year, month, data
'''
input_file = 'data_hmm.txt'
data = np.loadtxt(input_file, delimiter=',')

'''
Stack the data column wise for analsys
'''
# Arrange data for training 
X = np.column_stack([data[:,2]])


'''
Train HMM using 4 components.
'''
# Create and train Gaussian HMM 
print "\nTraining HMM...."
num_components = 4
model = GaussianHMM(n_components = num_components, 
                    covariance_type = "diag", 
                    n_iter = 1000)
model.fit(X)

# Predict the hidden states of HMM 
hidden_states = model.predict(X)

print "\nMeans and variances of hidden states:"
for i in range(model.n_components):
    print "\nHidden state", i+1
    print "Mean =", round(model.means_[i][0], 3)
    print "Variance =", round(np.diag(model.covars_[i])[0], 3)

# Generate data using model
num_samples = 1000
samples, _ = model.sample(num_samples) 
plt.plot(np.arange(num_samples), 
         samples[:,0], 
         c='black')

plt.title('Number of components = ' + str(num_components))

plt.show()
```

# Conditional Random Fields(discriminative/probabilistic model) #

Features: 

* CRFs are used to label and segment sequential data. 
* It is a discriminative or probabilistic model to analyze sequences, stocks, speech, words etc.
* They take as input - labeled observation sequence and it outputs a conditional probability distribution over this sequence.

Advantages of CRFs over HMM:

* **CRFs do not assume that the current output is independant of the previous outputs.**
* CRFs outperform HMMs in linguistics, bioinformatics, speech analysis.


```python
'''
Using CRFs to analyze a sequence of letters
'''

# pip install pystruct
# pip install cvxopt --user

# Necessary imports
import os
import argparse 
import cPickle as pickle 

'''
Using the load_letters dataset which consists of segmented letters and their associated feature vectors.
'''
from pystruct.datasets import load_letters
from pystruct.models import ChainCRF
from pystruct.learners import FrankWolfeSSVM

'''
Argument parser that takes c-value as a parameter. This parameter controls how specific we want our model to be without
losing the power to generalize.
'''
def build_arg_parser():
    parser = argparse.ArgumentParser(description='Trains the CRF classifier')
    parser.add_argument("--c-value", 
                        dest="c_value", 
                        required=False, 
                        type=float,
            default=1.0, help="The C value that will be used for training")
    return parser

'''
Class to handle CRF related processing.
'''
class CRFTrainer(object):
    '''
    init function takes in c value and classifier name
    '''
    def __init__(self, c_value, classifier_name='ChainCRF'):
        # Define c value
        self.c_value = c_value
        # Define classifier
        self.classifier_name = classifier_name

        if self.classifier_name == 'ChainCRF':
            model = ChainCRF()
            # FrankWolfeSSVM is a variation of the SVM model
            self.clf = FrankWolfeSSVM(model=model, 
                                      C=self.c_value, 
                                      max_iter=50) 
        else:
            raise TypeError('Invalid classifier type')
    
    
    # Load data and convert o numpy array for training and making predictions
    def load_data(self):
        # Load the dataset
        letters = load_letters()
        # X = data
        # y = labels
        X, y, folds = letters['data'], letters['labels'], letters['folds']
        # convert to numpy array
        X, y = np.array(X), np.array(y)
        return X, y, folds

    # X is a numpy array of samples where each sample
    # has the shape (n_letters, n_features) 
    def train(self, X_train, y_train):
        self.clf.fit(X_train, y_train)

    # Check performace
    def evaluate(self, X_test, y_test):
        return self.clf.score(X_test, y_test)

    # Run the classifier on input data
    def classify(self, input_data):
        return self.clf.predict(input_data)[0]

'''
Convert numbers back to alphabets
'''
def decoder(arr):
    alphabets = 'abcdefghijklmnopqrstuvwxyz'
    output = ''
    for i in arr:
        output += alphabets[i] 

    return output

if __name__=='__main__':
    
    args = build_arg_parser().parse_args()
    c_value = args.c_value

    crf = CRFTrainer(c_value)
    X, y, folds = crf.load_data()
    X_train, X_test = X[folds == 1], X[folds != 1]
    y_train, y_test = y[folds == 1], y[folds != 1]

    print "\nTraining the CRF model..."
    crf.train(X_train, y_train)

    score = crf.evaluate(X_test, y_test)
    print "\nAccuracy score =", str(round(score*100, 2)) + '%'

    print "\nTrue label =", decoder(y_test[0])
    predicted_output = crf.classify([X_test[0]])
    print "Predicted output =", decoder(predicted_output)
```

# Using HMM to create a generative model for the Yahoo stock # 

```python
import datetime

import numpy as np
import matplotlib.pyplot as plt
from hmmlearn.hmm import GaussianHMM
import pprint

'''
Import Yahoo stock data
'''
try:
    from matplotlib.finance import quotes_historical_yahoo_ochl
except ImportError:
    from matplotlib.finance import quotes_historical_yahoo as quotes_historical_yahoo_ochl

'''
Get quotes from Yahoo finance between 1994 and 2015. 

Data structure:

open value, closing value, high value, low value, volume
'''
quotes = quotes_historical_yahoo_ochl("INTC", 
        datetime.date(1994, 4, 5), datetime.date(2015, 7, 3))

pprint.pprint(quotes[0:5])

### DATA PREPROCESSING
'''
Extract the required values - closing value of stock, volume of stock, dates
'''
dates = np.array([quote[0] for quote in quotes], dtype=np.int)
closing_values = np.array([quote[2] for quote in quotes])
volume_of_shares = np.array([quote[5] for quote in quotes])[1:]

# Take diff of closing values and computing rate of change
diff_percentage = 100.0 * np.diff(closing_values) / closing_values[:-1]
dates = dates[1:]

# Stack the percentage diff and volume values column-wise for training
X = np.column_stack([diff_percentage, volume_of_shares])

### TRAINING HIDDEN MARKOV MODEL 
'''
Create and train Gaussian HMM 
'''
print "\nTraining HMM...."
model = GaussianHMM(n_components=5, 
                    covariance_type="diag", 
                    n_iter=1000)
# Fit using the stacked data - percentage difference, volume of stock
model.fit(X)

### GENERATE HMM MODEL OF THE YAHOO STOCK DATA
'''
Generate data using model
'''
num_samples = 500 
samples, _ = model.sample(num_samples) 
plt.plot(np.arange(num_samples), 
         samples[:,0], 
         c='black')
plt.title("Hidden Markov Model of Yahoo stock data")
plt.show()
```
